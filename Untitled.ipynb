# Description of the Dataset

The dataset (see gif file) is a 2D plot of the sea level pressure for each month between 1993 and 2020. So we can describe the signal as 

$SLP = SLP(t, x,y)$

How do we get the main variations of this signal ? The temporal variability ? 

Last week, we discuss Fourier transforms: it consist into decomposing a temporal serie into a sum of sinus with different periods and different amplitudes. Using this, we were able to get the main variations as the sinus with the largest amplitude (pic in fourier space). 

Remember that it requiered a periodic signal. Here, we have (i) a signal that is not periodic and (ii) a signal that varies in space 

--> Solution : Empirical Orthogonal Functions ! (EOFs) 

$SLP = SLP(t, x,y) = \sum_{i=0}^n{EOF_i(x,y) PC_i(t)}$

## Sans le cycle saisonnier

```python
ds['msl_cos'] = (ds.msl.groupby('time.month')-ds.msl.groupby('time.month').mean('time'))
ds['msl_cos'] = (ds.msl_cos-ds.msl_cos.mean('time'))*np.cos(ds.latitude*np.pi/180)
```

```python
model = EOF(ds.msl_cos, n_modes=15, norm=False, dim='time')
model.solve()
expvar = model.explained_variance_ratio()
eofs = model.eofs()
pcs = model.pcs()
eofs['latitude'] = ds.latitude
eofs['longitude'] = ds.longitude
PCS = xr.Dataset({'time':pcs.time.values, 'mode':pcs.mode.values})
PCS['pcs'] = pcs
```

```python
def running_mean(t, x, N) :
    rm = np.convolve(x, np.ones((N,))/N, mode='valid')
    times = t[int(N/2):len(rm)+int(N/2)]
    return times, rm
```

```python
N = 3
Na = 3*12
fig = plt.figure(figsize=(12,4*N))

for i in range(N):
    ax0 = fig.add_subplot(N,2,2*i+1)
    ax1 = fig.add_subplot(N,2,2*i+2, projection = ccrs.Robinson(central_longitude = -40))
    PCS.pcs.sel(mode=i+1).plot(ax=ax0)
    tm, rm = running_mean(PCS.time.values,  PCS.pcs.sel(mode=i+1).values, Na)
    ax0.plot(tm, rm,c = 'r')
    ax1.coastlines(color='.5')
    eofs.sel(mode=i+1).plot(ax=ax1, x = 'longitude', y = 'latitude', transform = ccrs.PlateCarree())
    ax1.coastlines()
    ax0.set_title('Var: ' +str(np.round(expvar.sel(mode = i+1).values*100, 3)) +r'$\%$')
    ax0.set_xlabel('')
    ax0.set_ylabel(r'$PC_%i(t)$'%i)
    ax1.set_title(r'$EOF_%i(x,y)$'%i)
plt.tight_layout()
```

```python
def change_time_coord(ds2):
    ds2 = ds2.rename({'time':'tdim', 'xgrid':'x', 'ygrid':'y'})
    return ds2
```

```python
dsI = xr.open_mfdataset('/storage/nplanat/DataNOAA_SEAICE/seaice_conc_monthly_nh_*', concat_dim = 'tdim', combine='nested', preprocess = change_time_coord)
dsI = dsI.drop({'melt_onset_day_cdr_seaice_conc_monthly', 'nsidc_bt_seaice_conc_monthly', 'nsidc_nt_seaice_conc_monthly',
         'qa_of_cdr_seaice_conc_monthly', 'stdev_of_cdr_seaice_conc_monthly'})
dsI['Month'] = (('tdim'), np.array([pd.Timestamp(dsI.tdim.values[i]).to_pydatetime().month for i in range(dsI.tdim.shape[0])]))
```

```python
dsIC = xr.Dataset(coords = {'x':dsI.x.values, 'y':dsI.y.values, 'year':np.arange(1979, 2020)})
SIC = np.nan*np.zeros((np.arange(1979, 2020).shape[0], dsI.y.shape[0], dsI.x.shape[0]))
for i in range(0, 2020-1979):
    SIC[i,:,:]=dsI.cdr_seaice_conc_monthly.isel(tdim = slice(11+i*12, 11+i*12+7)).mean('tdim', skipna = True).values
dsIC['SIC'] = (('year', 'y', 'x'), SIC)
```
